## TODO 2

This is the RAG pipeline. Here, we provide context for the LLM to provide grounded explanations.

This involves:
1. Providing the right context for the features
2. Search and Retrieval based on the Query (Hybrid Search)
3. Synthesising the model response based on the augmented query

```python
def semantic_chunker_to_faiss(cohere_api_key: str, 
                              text: str,
                              index_file_path: str="./data/agentic_rag/semantic_chunks.faiss",
                             ):

    # Step 1: Chunking Feature Descriptions into the Vector DB via Semantic Chuncking
    # Initialize the cohere client
    co = cohere.ClientV2(api_key=COHERE_API_KEY)
    encoder = CohereEncoder(name="embed-english-v3.0", 
                            cohere_api_key=COHERE_API_KEY) 
    
    # Initialize the semantic Chunker
    chunker = StatisticalChunker(
        encoder=encoder
    ) 

    # Semantically Chunking the text
    chunks = chunker(docs=[text.replace("--", "")])
    chunk_texts = [ chunk.content for chunk in chunks[0]]

    # Print the shunks
    chunker.print(chunks[0])
    # Save Chunk Texts
    with open("./data/agentic_rag/semantic_chunks.pkl", "wb") as f:
        pickle.dump(chunk_texts, f)

    # Step 2: Embed the chunks using Cohere
    response_embeds = CO.embed(
        texts=chunk_texts,
        model="embed-english-v3.0",
        input_type="search_document",  # Suitable for RAG/document embedding
        embedding_types=["float"]
    ).embeddings

    # response_embeds
    doc_embeds = np.array(response_embeds.float_, dtype=np.float32)

    # Step 3: Create the Search Index with FAISS
    emb_dim = doc_embeds.shape[1]
    index = faiss.IndexFlatL2(emb_dim)
    print(index.is_trained)
    index.add(doc_embeds)

    # Step 4: Save the FAISS index to disk
    faiss.write_index(index, index_file_path)
    print(f"\nFAISS index saved to {index_file_path}")



semantic_chunker_to_faiss(cohere_api_key=COHERE_API_KEY, text=text)
```

```python
def text_split_chunking(cohere_api_key:str, 
                       text: str,
                       index_file_path: str = "./data/agentic_rag/text_splitter_chunks.faiss"):

    # Chunking based on text structure
    co = cohere.ClientV2(api_key=COHERE_API_KEY)

    texts = text.split("--")

    # Embedding the text chunks
    response = CO.embed(
        texts=texts,
        model="embed-english-v3.0",
        input_type="search_document"
    ).embeddings

    doc_embeds = np.array(response.float_, dtype=np.float32)

    emb_dim = doc_embeds.shape[1]
    index = faiss.IndexFlatL2(emb_dim)
    print(index.is_trained)
    index.add(doc_embeds)

    # Step 4: Save the FAISS index to disk
    faiss.write_index(index, index_file_path)
    print(f"\nFAISS index saved to {index_file_path}")



# calling the function
text_split_chunking(
    cohere_api_key=COHERE_API_KEY,
    text=text,
)
```

```python
# Building the Dense Search pipeline

def semantic_search(query, 
                    number_of_results=4,
                    texts_path="./data/agentic_rag/semantic_chunks.pkl",
                    semantic_index_path="./data/agentic_rag/semantic_chunks.faiss"):
    # Load the semantic_chunk index
    semantic_index = faiss.read_index(semantic_index_path)
    # Get the query embedding
    query_embed = CO.embed(
        texts=[query],
        model="embed-english-v3.0",
        input_type="search_query",
        embedding_types=["float"],
    ).embeddings.float[0]

    # Retrieve nearest neighbors
    distances , similar_item_ids = semantic_index.search(np.float32([query_embed]), number_of_results)

    # 3. Format the results 
    # First retrieve the chunk texts
    with open(texts_path, "rb") as f:
        texts = pickle.load(f)
    texts_np = np.array(texts) # Convert texts list to numpy for easier indexing
    results = pd.DataFrame(data={'texts': texts_np[similar_item_ids[0]], 
                              'distance': distances[0]})
  
    # 4. Print and return the results
    print(f"Query:'{query}'\nNearest neighbors:")
    return results


query = "How about the emotional tone of the review?"

results = semantic_search(query)

results
```

```python
# Data set-up

##############

def run_agent(user_query, review_df, model="command-a-03-2025"):
    # Step 1: Get the user message
    messages = [
            {
                # System
                "role": "system",
                # System Instruction
                "content": """You are an XAI system designed for undergraduate students, generating clear, simple, and concise 
                grounded explanations using tools. Always use everyday language, avoid technical jargon (explain any necessary terms simply, 
                like "feature importance means how much a part of the review affects the AI's decision"), 
                and keep responses short—aim for 3-5 sentences or bullet points per explanation. Structure your answers with a 
                quick summary first (e.g., "The Review Agent sees this review as deceptive due to its word choices"), 
                1-2 easy details with analogies, and end with what it means for the user.
You are equipped with two tools for grounded explanations.
Plan steps: Retrieve feature importances using the get_prediction_context tool, and to provide grounded explanations based on the features use the text_splitter_search tool. For general questions about the review, 
ground explanations using the text_splitter_search tool. Use tools to get data, then synthesize a grounded, student-friendly response."""
            },
            # User content
            {
                "role": "user", 
                # User Query
                "content": user_query
            },
        ]
    
    
   
    
    # iteration = 0
    
    first_call = True
    
    # Step 2: Generate tool calls (if any)
    # Chat Params based on user message
    chat_params = {
        "model": model,
        "messages": messages,
        "tools": tools,
        "temperature": 0.1
        }
    if first_call:
         # Update User Content with the review DataFrame
        messages[1]["content"] += f"\nReview DataFrame: {review_df.to_dict(orient='records')}"
        chat_params["tool_choice"] = "REQUIRED"  # Force tool use; valid for Cohere
        first_call = False
    try:
        # Generate response without tools
        response = CO.chat(**chat_params)
    except Exception as e:
        return f"API Error: {str(e)}"
     
    while response.message.tool_calls:
        print("TOOL PLAN:")
        print(response.message.tool_plan, "\n")
        print("TOOL CALLS:")
        # Assistant for tool calls
        messages.append({
            "role": "assistant", 
            # "content": response.message.content[0].text, 
            "tool_calls": response.message.tool_calls,
            "tool_plan": response.message.tool_plan,    
            })
        tool_content = []
        for tool_call in response.message.tool_calls:
            # If running a semantic search
            if tool_call.function.name == "text_splitter_search":
                print(f"Tool Name: {tool_call.function.name}")
                # Get the query
                query = json.loads(tool_call.function.arguments)["query"]
                # Get the reasults from the tool call
                tool_results = text_splitter_search(query)
                # Append results
                tool_content.append(json.dumps(tool_results))
                # Append the tool output to the messages
                messages.append({
                    "role": "tool", 
                    "tool_call_id": tool_call.id, 
                    "content": tool_content
                })

            # elif running a prediction_context
            elif tool_call.function.name == "get_prediction_context":
                # Get the df
                review_df_dict, query_str = json.loads(tool_call.function.arguments)["review_df"], json.loads(tool_call.function.arguments)["query"]
                # Convert to a DF
                print(f" Review DF Dict{review_df_dict}")
                review_df = pd.DataFrame([review_df_dict])
                # print(f" Review DF -> {review_df}")
                # Get results from the prediction context
                tool_results = get_prediction_context(review_df=review_df, query=query_str)
                print(f" Tool Results --> {tool_results}")
                # Append results
                tool_content.append(str(tool_results))
                # Append the tool output to the messages
                messages.append({
                    "role": "tool", 
                    "tool_call_id": tool_call.id, 
                    "content": tool_content
                })
        
            
        # Step 4: Generate response
        response = CO.chat(model=model, 
                           messages=messages, 
                           tools=tools, 
                           temperature=0.3)
    
    messages.append({"role": "assistant", "content": response.message.content[0].text})
    # Print final response
    print("Response:")
    print(response.message.content[0].text)
    print("=" * 50)
    
    return messages



# Tests

df_url = "data/stages_data/stage_2_data.csv"

df = pd.read_csv(df_url)

review_df = df.drop(["mis_classified", "label"], axis=1).iloc[[0]]
    
user_query = "Based on the emotional appeal, why is this review deceptive?" 

run_agent(
    review_df=review_df,
    user_query=user_query
    
)
```
