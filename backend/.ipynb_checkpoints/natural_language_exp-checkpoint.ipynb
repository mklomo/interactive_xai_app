```python
import pandas as pd
import joblib
import cohere
from interpret.glassbox import ExplainableBoostingClassifier
import json

# ==================== CONFIGURATION ====================
# Replace with your actual Cohere API key (get it from https://dashboard.cohere.com/api-keys)
COHERE_API_KEY = "ZjORDZ9wI6rILugbfFjHpS2ZsCg9MKb3BRj9PtEx"
# Paths
INPUT_CSV = "./data/data_for_nl_exp_gen.csv"
MODEL_PATH = "./ml_training/ebm_classifier_trained_on_09_12_2025.joblib"
OUTPUT_CSV = './data/reviews_with_explanations.csv'
# ======================================================

# 1. Setup and Data Loading
co = cohere.ClientV2(api_key=COHERE_API_KEY)

# Load the dataset
print("Loading dataset...")
df = pd.read_csv(INPUT_CSV)

# Load your pre-trained EBM model
print("Loading EBM model...")
ebm = joblib.load(MODEL_PATH)

# Define the feature columns used during training
feature_cols = [
    'proportion_of_emotional_content_in_review',
    'proportion_of_adjectives_in_review',
    'readability_of_review',
    'analytic_writing_style'
]
X = df[feature_cols].copy()

# 2. Get Local Explanations from EBM (once for the whole dataset)
print("Generating local explanations for all reviews...")
ebm_local = ebm.explain_local(X)

def generate_nl_explanation(row_index, row_values):
    """
    Constructs a prompt and calls Cohere to generate a natural language explanation.
    Now includes the exact numerical feature contribution scores in BOTH the prompt
    and the generated explanation.
    """
    # Extract names and scores for this specific instance
    local_data = ebm_local.data(row_index)
    feature_names = local_data['names']
    feature_scores = local_data['scores']
   
    # Get the model's actual prediction
    prediction = ebm.predict(X.iloc[[row_index]])[0]
   
    # Build clean contribution dict (skip the intercept term) AND text for the prompt
    feature_contribs = {}
    contributions = []
    for name, score, value in zip(feature_names, feature_scores, row_values):
        if name.lower() == 'intercept':
            continue
        feature_contribs[name] = round(float(score), 4)
        direction = "pushed the prediction toward" if score > 0 else "pushed the prediction away from"
        contributions.append(f"- {name} (value: {value:.4f}) → contribution {score:.4f} ({direction} '{prediction}')")
   
    contribution_text = "\n".join(contributions)
   
    # Updated prompt: explicitly ask Cohere to include the actual scores in the explanation
    prompt = f"""You are an expert analyst explaining why an Explainable Boosting Machine (EBM) model classified a hotel review as genuine or deceptive.

Review text:
"{df.loc[row_index, 'review_text']}"

Model Prediction: {prediction}

Feature Contributions (these are the exact numerical scores from the model):
{contribution_text}

Write a concise (3-5 sentences), natural, professional explanation of WHY the model made this prediction.
You MUST mention the exact numerical contribution scores (e.g. "proportion_of_emotional_content_in_review contributed +0.1234") and tie them directly to the review text and the final prediction.
Speak like a helpful human analyst. Be insightful, easy to understand, and reference the specific scores where relevant."""

    # Call Cohere API (V2 client + latest model)
    response = co.chat(
        model="command-a-03-2025",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7,
        max_tokens=300
    )
   
    # Correct parsing for the response body you shared
    generated_text = response.message.content[0].text.strip()
   
    return generated_text, prediction, feature_contribs

# 3. Iterate through the dataframe and generate explanations + contributions
print(f"Generating natural-language explanations for {len(df)} reviews...")
explanations = []
model_predictions = []
feature_contribution_dicts = []

for i in range(len(df)):
    row_values = X.iloc[i].values
   
    print(f" → Processing review {i+1}/{len(df)}...")
    nl_exp, pred, contrib_dict = generate_nl_explanation(i, row_values)
   
    explanations.append(nl_exp)
    model_predictions.append(pred)
    feature_contribution_dicts.append(contrib_dict)

# 4. Create the final DataFrame
df['model_prediction'] = model_predictions
df['natural_language_explanation'] = explanations
df['feature_contributions'] = feature_contribution_dicts   # dict of feature → exact score

# Save the result
df.to_csv(OUTPUT_CSV, index=False)

print("\n✅ Process complete!")
print(f"   Final dataframe saved to '{OUTPUT_CSV}'")
print(f"   New columns added:")
print(f"     • 'model_prediction' (actual EBM prediction)")
print(f"     • 'natural_language_explanation' (Cohere-generated explanation that now includes the exact scores)")
print(f"     • 'feature_contributions' (Python dict with exact numerical contribution scores for each feature)")
```
